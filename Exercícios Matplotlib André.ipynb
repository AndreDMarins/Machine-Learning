## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

# Exercício 1

** Acompanhe estes passos: **
* ** Crie um objeto de figura chamado fig usando plt.figure () **
* ** Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax". **
* ** Plote (x, y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo: **

```python
#cria a figura em branco
fig = plt.figure()

#
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercício 2
** Crie um objeto de figura e coloque dois eixos sobre ele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente. **

```python
#criar a figura
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

```

** Agora plote (x, y) em ambos os eixos. E chame seu objeto de figura para mostrá-lo. **

```python
#criar a figura
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

#plotando da figura 1
ax1.plot(x,y,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Graficão')

#plotando na figura 2
ax2.plot(y,x,'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax1.set_title('Grafiquinho')
```

## Exercício 3

** Crie o gráfico abaixo, adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0,2,0,5, .4, .4] **

```python
fig = plt.figure()
ax3 = fig.add_axes([0,0,1,1])
ax4 = fig.add_axes([0.2,0.5,.4,.4])
```

** Agora use x, y e z arrays para recriar o gráfico abaixo. Observe os limites xlimits e y no gráfico inserido: **

```python
ax3.plot(x,z)
ax3.set_xlabel('X')
ax3.set_ylabel('Z')


ax4.plot(x,y)
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_title('zoom')
ax4.set_xlim(20,22)
ax4.set_ylim(30,50)

fig
```

## Exercício 4

** Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo. **

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

** Agora trace (x, y) e (x, z) nos eixos. Note a largura de linha e o estilo objetivos. **

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'b--',linewidth=3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('titulo')

axes[1].plot(x,y**2,'r', linewidth=4)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('titulo')

fig
plt.tight_layout()
```

** Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () apenas copiando e colando seu código anterior. **

```python

fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,3))

axes[0].plot(x,y,'b--',linewidth=3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('titulo')

axes[1].plot(x,y**2,'r', linewidth=4)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('titulo')

fig
plt.tight_layout()
```
